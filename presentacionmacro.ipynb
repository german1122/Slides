# Individuos heterogéneos en ingreso y Capital Humano con Incertidumbre

Este modelo es un replanteamiento al modelo de [Galor & Zeira (1993)](http://www.econ.nyu.edu/user/debraj/Courses/Readings/GalorZeira.pdf)

## Idea básica del modelo de Galor & Zeira

   **i)** La motivación es analizar qué papel juega la distribución del ingreso mediante la inversión en capital humano. 
   
   **ii)** Dos de los supuestos cruciales en este modelo es que los mercados de crédito son imperfectos y que la inversión en capital humano es indivisible. 
    
   **iii)** Los resultados a partir de estos dos supuestos "cruciales" indican  que la distribución inicial del ingreso afecta el producto agregado y la inversión, ambos en el corto y largo plazo (múltiples estados estacionarios). 
 
 

## Supuestos más importantes del modelo

**i)** Los individuos son altruistas y las generaciones de padres e hijos están traslapadas.

**ii)** Los individuos viven por dos periodos. En el primero pueden o no invertir en capital humano, de tal suerte que serán educados (skilled) o no en el segundo. Dado lo anterior, consumen y dejan una herencia o legado.  

**iii)** La imperfección en el mercado de crédito se obtiene de suponer que existen costos de garantía de pago 
y supervisión sobre los prestatarios.

**iv)** La inversión en capital humano es indivisible (los costos fijos de producir una unidad son muy altos, por lo que el bien es de difícil acceso a los consumidores). Es decir, la tecnología es no convexa, lo que resultará en equilibrios dinámicos de largo plazo no ergódicos (se asegura convergencia).

## El modelo teórico y el equilibrio
Las empresas no pueden evadir el pago por razones de reputación, por lo que piden prestado a una tasa $r$.
$$F_K(K_t, L_t^s) = r$$ 
Lo anterior determina al "ratio" entre capital y trabajo constante en $w_s$ en el mercado de los educados. Dos supuestos adicionales: los mercados de bienes y trabajo son perfectamente competitivos. 

Los prestamistas incurren en costos de garantía de pago y supervisión y la intermediación financiera es competitiva, pero los costos ya mencionados provocan que la tasa de interés sea mayor a la de retorno ($i=\frac{1+\beta r}{\beta-1} > r$).

## Cambios al modelo de Galor & Zeira
Cualquier cosa no especificada en lo siguiente se mantiene dentro de los supuestos:

**i)** En el periodo $t$ los padres deciden si prestar o no, de acuerdo con su riqueza inicial $x_t$.

**ii)** Ahora, las empresas siguen realizando una intermediación financiera, pero si un individuo incurre en impago parcial o total, entonces las empresas incurren en impago parcial o total del principal que deben pagar a sus prestamistas. Por tanto, si un individuo incurre en impago, daña de manera directa a los que de manera indirecta le prestaron. 
Es decir, $F_K(K_t, L_t^s) \neq r$.

**iii)** Los padres deciden si prestar o no, o pedir prestado o no (de acuerdo con la distribución de $x_t$) de acuerdo con su expectativa, a priori, sobre que quien paga el principal incurra o no en impago. 

$$ \mathbb{E}(x_{t+1}) =
  \begin{cases}
     (1-\alpha)[(x_t + w_n)(1+r) + w_n]P(w_s > (h-x_t)(1+i)), \quad x_t<f \\
 (1-\alpha)[(x_t -h)(1+i) + w_s]P(w_s > (h-x_t)(1+i)), \quad f\leq x_t<h\\
   (1-\alpha)[(x_t -h)(1+r) + w_s]P(w_s > (h-x_t)(1+i)), \quad f\leq h \leq x_t\\
  \end{cases}
$$


donde $\mathbb{E}(x_{t+1}) = b_j^k(x_t)P(w_s > (h-x_t)(1+i)), \forall j = n, s; k = 1, 2$.

## Resultados


Lo que esperamos hallar es que si incrementa la probabilidad de impago, entonces incrementa el número de personas que serán no educadas, dado que tendrán una perdidad de riqueza. Por el contrario, en caso de que dismiuya, esperamos que incremente la cantidad de individuos educados. 

* Uno de los supuestos que necesitaremos asegurar es la no-ergodicidad (mediante la indivisibilidad de la inversión en capital humano) y la monotonicidad creciente de la función de expectativas.
